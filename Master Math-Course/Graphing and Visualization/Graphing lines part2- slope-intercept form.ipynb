```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [-5, 5]
m = 2
b = 1

# y = [340, 2345]

# for i in range(0, len(x)):
#     y[i] = m*x[i] + b

y = m*np.array(x) + b
    

plt.plot(x, y,label='y=%sx+%s' %(m, b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```

```python
#1)line y = 0.7x + - 2
#2)line y=-1.25x + 0.75

x = [-5, 5]
m = [.7, -1.25]
b = [-2, .75]


for i in range(0, len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x, y, label='y=%sx+%s' %(m[i], b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
plt.show()
```
