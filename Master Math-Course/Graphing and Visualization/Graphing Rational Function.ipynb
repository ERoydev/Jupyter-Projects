```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = range(-3, 4)
y = np.zeros(len(x))

for i in range(0, len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y, 's-')
plt.grid()

plt.show()    
```

```python
x = np.linspace(-3, 4, 100)

y = 2 + np.sqrt(abs(x))

plt.plot(x, y, 'm-')
plt.show()
```

```python
x = np.linspace(-4, 4, 200)
e = range(-1, 4)

for i in e:
    y = x**i
    plt.plot(x, y,label='$y=x^{%s}$' %i, linewidth=4)
    
plt.legend()
plt.xlim(x[0], x[-1])
plt.ylim([-20, 20])
plt.show()
```
